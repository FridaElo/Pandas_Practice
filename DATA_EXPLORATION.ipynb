```python
world = pd.read_csv('world_population.csv')
world
```

```python
sns.heatmap(world.corr(), annot = True)
plt.rcParams['figure.figsize']=(25,7)
plt.show
```

```python
df2 = world.groupby('Continent')[['1970 Population','1980 Population','1990 Population','2000 Population','2010 Population','2015 Population','2020 Population','2022 Population']].mean().sort_values(by='2022 Population', ascending = False)
df2
```
